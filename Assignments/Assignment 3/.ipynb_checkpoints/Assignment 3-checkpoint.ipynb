```python
import argparse
import cv2
refPt = []
Path1 = "./images/image11.png"
Path2 = "./images/image12.png"
Toggle = True
image = cv2.imread(Path1)
Overal = []
counter = 0
def Click(event, x, y, flags, param):
    global refPt, Toggle, image, Overal, counter
    if event == cv2.EVENT_LBUTTONDOWN:
        refPt = [x, y, 1]
        Overal.append(refPt)
        Toggle = not Toggle
        if Toggle:
            counter += 1
            image = cv2.imread(Path1)
        else:
            image = cv2.imread(Path2)
        

clone = image.copy()
cv2.namedWindow("image")
cv2.setMouseCallback("image", Click)

while True:
    cv2.imshow("image", image)
    key = cv2.waitKey(1) & 0xFF
    if key == ord("c") or key == ord("C") or key == ord("q") or key == ord("Q") or counter == 10:
        break
my_string = "C = [\n"
for i in range(0,len(Overal)):
    
    if i % 2 == 0:
        my_string+= " [np.array([" + str(Overal[i][0]) + ", " + str(Overal[i][1]) + ", " + str(Overal[i][2]) + "]), np.array([" + str(Overal[i+1][0]) + ", " + str(Overal[i+1][1]) + ", " + str(Overal[i+1][2]) + "])]"
        if i != 18:
            my_string += ",\n"
        else:
            my_string += "\n"
my_string+= "]"
print(my_string)
cv2.destroyAllWindows()
print("Done!")
```
